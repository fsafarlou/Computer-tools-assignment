(l) As you know, a magic square is a matrix all of whose row sums,
column sums and the sums of the two diagonals are the same.
(One diagonal of a matrix goes from the top left to the bottom
right, the other diagonal goes from top right to bottom left.)
Show by direct computation that if the matrix A is given by:
A=np.array([[17, 24, 1, 8, 15],
[23, 5, 7, 14, 16],
[ 4, 6, 13, 20, 22],
[10, 12, 19, 21, 3],
[11, 18, 25, 2, 9]])
The matrix A has 5 row sums (one for each row), 5 column sums
(one for each column) and two diagonal sums. These 12 sums
should all be exactly the same, and you could verify that they
are the same by printing them and “seeing” that they are the
same. It is easy to miss small differences among so many numbers, though. Instead, verify that A is a magic square by constructing the 5 column sums and computing the maximum and
minimum values of the column sums. Do the same for the 5 row
sums, and compute the two diagonal sums. Check that these
six values are the same. If the maximum and minimum values
are the same, the flyswatter principle says that all values are the
same. Hints: The function np.diag extracts the diagonal of a matrix, and the function np.fliplr extracts the other diagonal.

```python
A = np.array([[17, 24, 1, 8, 15], 
             [23, 5, 7, 14, 16],
              [4, 6, 13, 20, 22],
              [10, 12, 19, 21, 3],
              [11, 18, 25, 2, 9]])
print(A.sum(axis=1))


print ("sum of row 1 = ", A[0].sum())
print ("sum of row 2 = ", A[1].sum())
print ("sum of row 3 = ", A[2].sum())
print ("sum of row 4 = ", A[3].sum())
print ("sum of row 5 = ", A[4].sum())
print ("sum of column 1 = ", A[:, 0].sum())
print ("sum of column 2 = ", A[:, 1].sum())
print ("sum of column 3 = ", A[:, 2].sum())
print ("sum of column 4 = ", A[:, 3].sum())
print ("sum of column 5 = ", A[:, 4].sum())
print ("sum of the diagonal = ", np.diag(A).sum())
print ("sum of the reversed diagonal = ", np.fliplr(A).diagonal().sum())
#works:)
```

(a)Use matplotlib.pyplot.plot to produce a plot of the functions
f(x) = e−x/10sin(πx) and g(x) = xe−x/3 over the interval [0,10]. Include labels for the x- and y-axes, and a legend explaining 3which line is which plot. Save the plot as a .jpg (“Jpeg”) file and send me a copy with your work.

```python
import matplotlib.pylab as plt
%matplotlib inline
x = np.linspace(0, 10, 1000)
pi= np.pi
f= np.exp((-x)/10)*np.sin(pi*x)
g= x* np.exp((-x)/3)

plt.title('Funciones')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.plot(x, f, color="forestgreen", label="f (x)=e^-x/10 sin(πx)", marker='+')
plt.plot(x,g , color="red", label="g(x) = xe−x/3")
plt.axhline(0, color="gray", linestyle="--")
plt.legend();
plt.savefig('functions3a.jpg')
```

(b) The shape of a limaçon can be defined parametrically as
r = r0 + cosθ
x = rcosθ
y = rsinθ
When r0 = 1, this curve is called a cardioid. Use this definition to
plot the shape of a limaçon for r0 = 0.8, r0 = 1.0, and r0 = 1.2. Be
sure to use enough points that the curve is closed and appears
smooth (except for the cusp in the cardioid). Use a legend to
identify which curve is which. Save the plot as a .pdf file and
send me a copy with your work.

```python
#%pylab inline
def limacon(r0):
    theta = np.linspace(0, 2 * pi, 1000)
    r=r0+cos(theta)
    x=r*cos(theta)
    y=r*sin(theta)
    plt.plot(x,y)
plt.xlabel('X-AXIS',size=20,color='green')
plt.ylabel('Y-AXIS',size=20,color='green')
plt.title('limacon',size=20,color='green')

    
a=limacon(0.8)
b=limacon(1.0)
c=limacon(1.2)
a= mpatches.Patch(color='blue', linestyle='--', label='limacon(0.8)')
b= mpatches.Patch(color='orange', linestyle='--', label='limacon(1.0)')
c= mpatches.Patch(color='green', linestyle='--', label='limacon(1.2)')
plt.legend(handles=[a,b,c])

plt.savefig('limacon.pdf')
```

9 Create a sequence of 10 images in which the sine function slowly evolves over time. Update the data in
the figure, do NOT create a new figure for each time value

```python
oldsize = pylab.rcParams['figure.figsize']
pylab.rcParams['figure.figsize'] = (2.,2)
pylab.figure()
x = np.arange(0,10.,0.1)
y = np.sin(x)
time_steps = np.linspace(0,10,10)

xs = x[x<time_steps[0]]
ys = y[x<time_steps[0]]
l,=pylab.plot(xs,ys,'k')
m,=pylab.plot(xs,ys,'or')
pylab.yticks([])
pylab.xlim(0,10)
pylab.ylim(-1,1)
pylab.show()
for i in range(1,10):
    pylab.xlim(0,10)
    pylab.ylim(-1,1)
    xs = x[x<time_steps[i]]
    ys = y[x<time_steps[i]]
    l.set_data(xs,ys)
    m.set_data(xs[-1],ys[-1])
    pylab.plot(xs,ys)
    pylab.show()
print
pylab.rcParams['figure.figsize'] = oldsize
    
```
